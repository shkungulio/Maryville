## A step-by-step walkthrough of an annotated plot (using matplotlib)



### The plot above looks very diffult to construct; but we will step through it a little at a time.

### Import the necessary packages. In this case, all we will need are numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Get (or create) the data. Here, we can generate the data by using numpy's linspace function, which let's you define two endpoints, and the number of equally spaced points you want between them. Then, we'll set the arrays C and S respectively, to the cosine and sine of these points.

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

### Instantiate the figure. It is a good housekeeping idea to set your figure size and give it a name as a subplot. this short name can then be used in the code referneces to this subfigure.

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
```

### The base plot is generated with just two lines of code:

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()

# Here is the new code
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet
```

### We will extend the x and y limits by 50%

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet

# Here is the new code
# change the x and y limit
plt.xlim(X.min()*1.5, X.max()*1.5)
plt.ylim(C.min()*1.5, C.max()*1.5)
```

### Next, we will recast the tick marks to fit better with these trigonometric curves. Note the use of \pi for the pi symbol.

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet
# change the x and y limit
plt.xlim(X.min()*1.5, X.max()*1.5)
plt.ylim(C.min()*1.5, C.max()*1.5)

# Here is the new code
# change the ticks
# ticks are just a way to 'change the values' represented on the x and y axis
plt.xticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'] # r before each string means raw string
)

plt.yticks(
    [-1, +1],
    [r'$-1$', r'$+1$']
)
```

### Next, let's remove the upper and right container lines since they do not add to the value of the plot. These are called splines. We change their color to 'none'.

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet
# change the x and y limit
plt.xlim(X.min()*1.5, X.max()*1.5)
plt.ylim(C.min()*1.5, C.max()*1.5)

# change the ticks
# ticks are just a way to 'change the values' represented on the x and y axis
plt.xticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'] # r before each string means raw string
)

plt.yticks(
    [-1, +1],
    [r'$-1$', r'$+1$']
)

# Here is the new code
# removes the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
```

### Our x-axis is still not where we would like it, so let's reposition it. At the same time, it does not need to be so dark, so lighten it to keep the focus on the curves, not the axis.

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet
# change the x and y limit
plt.xlim(X.min()*1.5, X.max()*1.5)
plt.ylim(C.min()*1.5, C.max()*1.5)

# change the ticks
# ticks are just a way to 'change the values' represented on the x and y axis
plt.xticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'] # r before each string means raw string
)

plt.yticks(
    [-1, +1],
    [r'$-1$', r'$+1$']
)
# removes the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Here is the new code
# changes the position of the other spines
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['bottom'].set_color('black') # this helps change the color
ax.spines['bottom'].set_alpha(.3) # and adds some transparency to the spines
```

### Similarly, we will move the y-axis to the zero point on the x-axis, and make it lighter as well.

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet
# change the x and y limit
plt.xlim(X.min()*1.5, X.max()*1.5)
plt.ylim(C.min()*1.5, C.max()*1.5)

# change the ticks
# ticks are just a way to 'change the values' represented on the x and y axis
plt.xticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'] # r before each string means raw string
)

plt.yticks(
    [-1, +1],
    [r'$-1$', r'$+1$']
)
# removes the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# changes the position of the other spines
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['bottom'].set_color('black') # this helps change the color
ax.spines['bottom'].set_alpha(.3) # and adds some transparency to the spines



# Here is the new code
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_color('black')
ax.spines['left'].set_alpha(.3)
```

### Here is some cool stuff: we will add an annotation at the point 2 pi / 3 on the x-axis. This will be done by adding a dashed vertical red line up to the sine curve (from the x-axis) and an annotation plotted at the point of intersection.

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet
# change the x and y limit
plt.xlim(X.min()*1.5, X.max()*1.5)
plt.ylim(C.min()*1.5, C.max()*1.5)

# change the ticks
# ticks are just a way to 'change the values' represented on the x and y axis
plt.xticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'] # r before each string means raw string
)

plt.yticks(
    [-1, +1],
    [r'$-1$', r'$+1$']
)
# removes the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# changes the position of the other spines
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['bottom'].set_color('black') # this helps change the color
ax.spines['bottom'].set_alpha(.3) # and adds some transparency to the spines

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_color('black')
ax.spines['left'].set_alpha(.3)


# Here is the new code
# annotate different values
t = 2*np.pi/3
# plot a straight line to connect different points
plt.plot([t, t], [0, np.sin(t)], color ='red', linewidth = 1.5, linestyle = "--", alpha = 0.5)
plt.scatter(t, np.sin(t), 50, color ='red', alpha = 0.5)
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$' + "\nNotice the the arrows are optional",
             xy = (t, np.sin(t)), 
             xycoords = 'data',
             fontsize = 16)
```

### And the coolness continues: We will do the same for the cosine curve, but with a green dashed line, and an arced arrow pointing to the intersection point. Also, we will add our title and legend. The resultant graph is attractive and easy to understand.

```python
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot()
ax.plot(X,S, color = "red", alpha = 0.5, lw = 3, label = "Sine")
ax.plot(X,C, color = "green", alpha = 0.5, lw = 3, label = "Cosine") # note that our labels will not show yet
# change the x and y limit
plt.xlim(X.min()*1.5, X.max()*1.5)
plt.ylim(C.min()*1.5, C.max()*1.5)

# change the ticks
# ticks are just a way to 'change the values' represented on the x and y axis
plt.xticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'] # r before each string means raw string
)

plt.yticks(
    [-1, +1],
    [r'$-1$', r'$+1$']
)
# removes the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# changes the position of the other spines
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['bottom'].set_color('black') # this helps change the color
ax.spines['bottom'].set_alpha(.3) # and adds some transparency to the spines

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_color('black')
ax.spines['left'].set_alpha(.3)

# annotate different values
t = 2*np.pi/3
# plot a straight line to connect different points
plt.plot([t, t], [0, np.sin(t)], color ='red', linewidth = 1.5, linestyle = "--", alpha = 0.5)
plt.scatter(t, np.sin(t), 50, color ='red', alpha = 0.5)
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$' + "\nNotice the the arrows are optional",
             xy = (t, np.sin(t)), 
             xycoords = 'data',
             fontsize = 16)



# Here is the new code
# do the same for cosine
plt.plot([t, t], [0, np.cos(t)], color = 'green', linewidth = 1.5, linestyle = "--", alpha = 0.5)
plt.scatter(t, np.cos(t), 50, color = 'green', alpha = 0.5)
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy = (t, np.cos(t)), 
             xycoords = 'data',
             xytext = (t/2, -1), 
             fontsize = 16,
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

# adjust the x and y ticks
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None', alpha = 0.65 ))
    
# add the title and a legend
plt.title("Sine and Cosine functions (Original by N. Rougier)")
plt.legend(loc = "upper left", fontsize = 12);
```
