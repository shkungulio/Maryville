```python
# Import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Problem 1: Working with 2D Arrays

In this task, I will generate a 5x8 array with randomly sampled elements and then perform specific operations on this array using NumPy. The operations include creating the array, rounding its elements, printing specific rows and columns, and calculating sums. The entire process will be carried out without using loops.

```python
# Set a seed for reproducibility
np.random.seed(1)

# Create a 5x8 array with elements sampled uniformly from the interval [0, 10]
Z = np.random.uniform(0, 10, (5, 8))

# Round the elements of Z to two decimal places
Z = np.round(Z, 2)

# Print the array Z
print("Display the array Z")
print("-"*43)
print(Z)
```

```python
row_3 = Z[2, :]
col_6 = Z[:, 5]

# Display Row 3 and Column 6 arrays
print("Display Row 3 and Column 6 Arrays")
print("-"*51)
print(f"{'Row 3:':<9} {row_3}")
print(f"Column 6: {col_6}")
```

```python
# Calculate the column sum and row sum for Z, as well as the sum of the entire array
row_sums = np.sum(Z, axis=1)
col_sums = np.sum(Z, axis=0)
array_sum = np.sum(Z)

# Display the sums
print("Display Row sums, Column sums, and Array sum")
print("-"*51)
print(f"{'Row Sums:':<12} {row_sums}")
print(f"{'Column Sums:':<2} {col_sums}")
print(f"Array Sum: {array_sum:>8.2f}")
```

## Problem 2: Reshaping and Stacking Arrays

In this problem, I will use NumPy to perform stacking and reshaping operations to create a 
1000 Ã— 4 array. Each column of the array will be drawn from a normal distribution with specified parameters. 

```python
# Set a seed of 167
np.random.seed(167)

# Create arrays x1, x2, x3, and x4
x1 = np.random.normal(50, 10, 1000)
x2 = np.random.normal(20, 5, 1000)
x3 = np.random.normal(100, 30, 1000)
x4 = np.random.normal(10, 2, 1000)

# Reshape to column arrays and combine into 1000 x 4 array
X = np.hstack((x1.reshape(-1, 1), x2.reshape(-1, 1), x3.reshape(-1, 1), x4.reshape(-1, 1)))

# Round the values to two decimal places
X = np.round(X, 2)

# Print the shape of X
shape_of_X = X.shape
print("The shape of X:", shape_of_X)
```

```python
# Display the shape of the first 6 rows of X
print("Elements in the first six rows are:")
print("-"*35)
print(X[:6])
```

## Problem 3: Standardization

In this problem, I will use numpy operations to create a scaled version of the array created in Problem 2. Specifically, we will perform standard scaling, where each column is rescaled to have a mean of 0 and a standard deviation of 1. This involves subtracting the mean of each column and then dividing by the standard deviation of that column.

```python
# Calculate column means and standard deviations
column_means = np.mean(X, axis=0)
column_stds = np.std(X, axis=0)

# Round the column_means and column_stds to two decimal places
#column_means = np.round(column_means, 2)
#column_stds = np.round(column_stds, 2)

# Print the results of column means and standard deviations
#print(f"{'Column means:':<27} {column_means}")
print(f"{'Column means:':<27} {' '.join(f'{means:.2f}' for means in column_means)}")
#print(f"Column standard deviations: {column_stds}")
print(f"{'Column standard deviations:':<} {' '.join(f'{std:.2f}' for std in column_stds)}")
```

```python
# Perform standardization
W = np.array((X - column_means) / column_stds)

# Calculate column means and standard deviations for W
W_means = np.mean(W, axis=0)
W_stds = np.std(W, axis=0)

# Print the results of column means and standard deviations for W
print(f"{'Column means:':<27} {' '.join(f'{means:.2f}' for means in W_means)}")
print(f"{'Column standard deviations:':<} {' '.join(f'{std:.2f}' for std in W_stds)}")
```

## Problem 5: Regional Counts and Means

Each automobile in this dataset was manufactured in one of three regions: Asia, Europe, or The United States. In this analysis, I will calculate the number of observations associated with each region and the regional means for mpg (miles per gallon), cyl (number of cylinders), and wt (weight).

```python
# Obtain an array of unique values in the 'region' column
regions = np.unique(auto['region'])
print(regions)
```

```python
# Create DataFrames for each region
asia_auto = auto[auto['region'] == 'asia']
eur_auto = auto[auto['region'] == 'eur']
usa_auto = auto[auto['region'] == 'usa']

# Number of cars in each region
num_asia = asia_auto.shape[0]
num_eur = eur_auto.shape[0]
num_usa = usa_auto.shape[0]

# Print the results
print(f"Number of cars manufactured in Asia: {num_asia:>4}")
print(f"Number of cars manufactured in Europe: {num_eur}")
print(f"Number of cars manufactured in USA: {num_usa:>6}")
```

```python
# Calculate the means for each region
asia_means = asia_auto[['mpg', 'cyl', 'wt']].mean(axis=0, numeric_only=True)
eur_means = eur_auto[['mpg', 'cyl', 'wt']].mean(axis=0, numeric_only=True)
usa_means = usa_auto[['mpg', 'cyl', 'wt']].mean(axis=0, numeric_only=True)

# Create a DataFrame for the means
mean_df = pd.DataFrame({
    'mpg': [asia_means['mpg'], eur_means['mpg'], usa_means['mpg']],
    'cyl': [asia_means['cyl'], eur_means['cyl'], usa_means['cyl']],
    'wt': [asia_means['wt'], eur_means['wt'], usa_means['wt']]
}, index=regions)

print(mean_df)
```

## Problem 6: Average Weight and MPG by Region

In this problem, I will visually represent the results for MPG and weight from Problem 5 in the form of bar plots.

```python
# Create a list of colors
colors1 = ['skyblue', 'lightgreen', 'salmon']

# Create the figure and bar plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

# Bar plot for average MPG by region
ax1.bar(mean_df.index, mean_df['mpg'], color=colors1, edgecolor='black')
ax1.set_title('Average MPG by Region')
ax1.set_xlabel('Region')
ax1.set_ylabel('Average MPG')

# Bar plot for average weight by region
ax2.bar(mean_df.index, mean_df['wt'], color=colors1, edgecolor='black')
ax2.set_title('Average Weight by Region')
ax2.set_xlabel('Region')
ax2.set_ylabel('Average Weight in Pounds')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

## Problem 7: Relationship between Weight and Miles Per Gallon

In this problem, I will create a scatter plot for each region displaying the relationship between MPG and weight for vehicles manufactured in each region.

```python
# Set the figure size
plt.figure(figsize=[12, 4])

# Colors used for the regions in the bar charts from Problem 6
region_colors = {
    'usa': 'salmon',
    'asia': 'skyblue',
    'eur': 'lightgreen'
}

# Loop over the regions and create subplots
for i, region in enumerate(regions):
    
    plt.subplot(1, 3, i + 1)
    
    # Boolean masking to select the relevant records
    region_data = auto[auto['region'] == region]
    
    # Scatter plot
    plt.scatter(region_data['wt'], region_data['mpg'], edgecolor='black', color=region_colors[region], alpha=0.8)
    
    # Set limits and labels
    plt.xlim([1200, 5000])
    plt.ylim([0, 50])
    plt.xlabel('Weight in Pounds')
    plt.ylabel('Miles Per Gallon')
    plt.title(f'Weight vs MPG ({region.capitalize()})')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

## Problem 8: Cylinder Distribution by Region

In this problem, I will create a stacked bar chart displaying the distribution of the number of cylinders in vehicles manufactured in each region.

```python
# Use np.unique() to create a list of unique values appearing in the cyl column
cyl_values = np.unique(auto['cyl'])
print("cyl_values:", cyl_values)
```

```python
# Create a DataFrame with one row for each value of cyl and one column for each value of region
cyl_counts_by_region = pd.crosstab(auto['cyl'], auto['region'])
cyl_counts_by_region
```

```python
# Convert the count information into proportions
cyl_props_by_region = cyl_counts_by_region.div(cyl_counts_by_region.sum(axis=0), axis=1)
#print(cyl_props_by_region)

# Calculate the bar bottoms
bar_bottoms = np.cumsum(cyl_props_by_region, axis=0) - cyl_props_by_region

# Create a list of colors
colors2 = ['red', 'blue', 'green', 'orange', 'purple']

# Create a Matplotlib figure
plt.figure(figsize=[6, 4])

# Loop over the rows of cyl_props_by_region to add a bar chart to the figure
for i, (cyl, row) in enumerate(cyl_props_by_region.iterrows()):
    plt.bar(row.index, row, bottom=bar_bottoms.loc[cyl], color=colors2[i], edgecolor='black', label=str(cyl))

# Set the labels and title
plt.xlabel('Region')
plt.ylabel('Proportion')
plt.title('Distribution of Cylinder Numbers by Region')

# Add a legend
plt.legend(title='Cylinders', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the figure
plt.show()
```
