```sql
SELECT count(actor_id) AS "Count of Actors"
FROM actor;
```

```sql
SELECT COUNT(DISTINCT concat(first_name, last_name)) AS "Unique Actors"
FROM actor;
```

```sql
SELECT *
FROM customer;
```

```sql
SELECT COUNT(customer_id) AS "Active Customers"
FROM customer
GROUP BY activebool
HAVING activebool = FALSE;
```

```sql
SELECT COUNT(customer_id) AS "Active Customers"
FROM customer
GROUP BY active
HAVING active = 1;
```

```sql
SELECT EXTRACT(month from payment_date)::int AS "Month"
    , MAX(amount) AS "Maximum Payment"
FROM payment
GROUP BY "Month"
ORDER BY "Month";
```

```sql
SELECT EXTRACT(month from payment_date)::int AS "Month"
    , SUM(amount) AS "Maximum Payment"
FROM payment
GROUP BY "Month"
HAVING EXTRACT(month from payment_date)::int = 5;
```

```sql
SELECT ROUND(AVG(amount),2) AS "Average Rental"
FROM payment;
```
