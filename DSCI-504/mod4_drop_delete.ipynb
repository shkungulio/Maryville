Your results should show that course\_id 2 is removed. If not, go back and look at your SQL statement. To delete records form a table with a foreign key constraint, you need to ensure you use a cascading query.

To delete rows in course where there is a faculty key constraint we have to reference the parent table's constraint columns. In this case, the course faculty ID column is a foreign key to the faculty table. We can delete the records that meet our criteria by passing the conditions in a USING clause.

> ```
> DELETE FROM course USING faculty
> ```

The first part of the statement identifies that we want to DELETE [rows] FROM course USING the following condition:

> ```
> WHERE course_professor = faculty.fac_id AND faculty.fac_status = 'No';
> ```

The conditions recreates a JOIN where we compare the professor in the course to the parent faculty ID colukn and the faculty's status. If it is False/'No' then we DELETE [row].
