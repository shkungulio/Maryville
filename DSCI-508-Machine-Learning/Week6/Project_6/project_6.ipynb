## Load Libraries and Dataset

```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
mpg_df = pd.read_csv('mpg.csv')

# Display the first few rows of the DataFrame to understand its structure
mpg_df.head()
```

## Create the Plots

### Stacked Histogram of Engine Displacement Colored by Year
To creates and displays a stacked histogram using Seaborn and Matplotlib. The histogram visualizes the distribution of engine displacement (displ) across different years (year), with separate colors for each year.

```python
# Plot histogram figure
plt.figure(figsize=(10, 6)) #creates a new figure for plotting with a specified size
sns.histplot(data=mpg_df, x='displ', hue='year', multiple='stack', palette='tab10', bins=30) #use Seaborn’s histplot function to create a histogram
plt.title('Stacked Histogram of Engine Displacement by Year') #sets the title of the histogram plot.
plt.xlabel('Engine Displacement') #labels the x-axis of the plot as "Engine Displacement"
plt.ylabel('Frequency') #labels the y-axis of the plot as "Frequency"
plt.tight_layout() #adjusts the layout of the plot to make sure that everything fits within the figure area and there is no overlap
plt.show() #displays the plot
```

### Violin Plot of Highway MPG Colored by Year
Using matplotlib and seaborn to create and display a violin plot of Highway MPG by year.

```python
# Plot Violin figure
plt.figure(figsize=(10, 6)) # initialize a new figure with a size of 10 inches wide and 6 inches tall
sns.violinplot(data=mpg_df, x='year', y='hwy', hue='year', inner='quartile', palette='tab10') #use Seaborn’s violinplot function to create a violin
plt.title('Violin Plot of Highway MPG by Year') #sets the title of the plot to "Violin Plot of Highway MPG by Year"
plt.xlabel('Year') #labels the x-axis of the plot as "Year"
plt.ylabel('Highway MPG') #labels the y-axis of the plot as "Highway MPG"
plt.tight_layout() #adjusts the layout of the plot to make sure that everything fits within the figure area and there is no overlap
plt.show() #displays the plot
```

### Scatter Plot of City vs Highway MPG with Regression Lines
To create a scatter plot of city versus highway mileage from the mpg_df DataFrame including regression lines for each year to show trends.

```python
# Plot scatter with regression line
plt.figure(figsize=(10, 6)) #creates a new figure for plotting with a specified size
#use Seaborn's lmplot function to create a scatter plot of the cty (city mileage) versus hwy (highway mileage)
sns.lmplot(data=mpg_df, x='cty', y='hwy', hue='year')
plt.title('Scatter Plot of City vs Highway MPG with Regression Lines') #sets the title of the plot.
plt.xlabel('City') #labels the x-axis of the plot as "City"
plt.ylabel('Highway') #labels the y-axis of the plot as "Highway"
plt.tight_layout() #adjusts the layout of the plot to make sure that everything fits within the figure area and there is no overlap
plt.show() #displays the plot
```
